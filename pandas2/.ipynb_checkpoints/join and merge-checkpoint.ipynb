```python
import pandas as pd
```

```python
secA = pd.DataFrame({
    'name' : ['raj', 'rohan', 'amar', 'khushi', 'yash'],
    'grade' : ['A', 'B', 'C', 'B', 'A'],
    'roll-no': [23, 45, 67, 27, 77]
})

secB = pd.DataFrame({
    'name' : ['rahul', 'ram', 'amit', 'khushboo'],
    'grade' : ['D', 'B', 'A', 'B'],
    'roll-no': [33, 15, 7, 47]
})
```

***Write a Pandas program to join the two given dataframes along rows and assign all data***

```python
res = pd.concat([secA, secB], ignore_index = True)
res
```

***Write a Pandas program to join the two given dataframes along columns and assign all data***

```python
res = pd.concat([secA, secB], axis = 1)
res
```

```python
left = pd.DataFrame({'Key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']},
                    index=['K0', 'K1', 'K2', 'K3'])
 
right = pd.DataFrame({'Key': ['K0', 'K1', 'K2', 'K3'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']},
                     index=['K0', 'K1', 'K2', 'K3'])
```

```python
merged = pd.merge(left, right, how= 'left', on=['Key'])
merged
```
